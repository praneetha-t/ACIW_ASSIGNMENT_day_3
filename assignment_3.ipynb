```python
import pandas as pd
data = pd.read_csv(r"c:\Users\sudhakar\Downloads\archive.zip")
```

```python
import matplotlib.pyplot as plt
X= data['Final_Percentage']
Y= data['Pass_Fail']
plt.scatter(x = X,y=Y,marker='*',c='green', alpha=0.8,linewidths=2,edgecolors='blue')
plt.scatter(x = X,y = data['Final_Percentage'], marker='o', c='red', alpha=0.2,linewidths=8)
plt.title("Final_Percentage v/s Pass_Fail")
plt.xlabel("Final_Percentage")
plt.ylabel("Pass_Fail")
plt.grid()
plt.show()
```
